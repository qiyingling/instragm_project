```python
import pandas as pd
import numpy as np
from faker import Faker
import random

faker = Faker()
```

```python
n_users= 10000
```

```python
users = []
for i in range(1, n_users + 1):
    users.append({
        "user_id": i,
        "user_name": faker.user_name(),
        "register_date":faker.date_between(
            start_date="-2y",
            end_date="today")
    })
users_df = pd.DataFrame(users)
users_df.head()
```

```python
users_df.to_csv("raw_users.csv",index=False)
```

```python
n_photos = 50000
photos=[]
for i in range(1,n_photos + 1):
    photos.append({
        "photo_id": i,
        "user_id": random.randint(1,n_users),
        "upload_time": faker.date_time_this_year()
    })
photos_df = pd.DataFrame(photos)
photos_df.head()
```

```python
photos_df.to_csv("raw_photos.csv",index=False)
```

```python
likes_df = pd.DataFrame({
    "user_id": np.random.randint(1, n_users + 1, 300000),
    "photo_id": np.random.randint(1, n_photos + 1, 300000),
    "like_time": [faker.date_time_this_year()
                  for _ in range(300000)]
})
likes_df.head()
```

```python
likes_df.to_csv("raw_likes.csv",index=False)
```

```python
comments_df = pd.DataFrame({
    "comment_id": range(1,100001),
    "user_id": np.random.randint(1, n_users + 1, 100000),
    "photo_id": np.random.randint(1, n_photos + 1, 100000),
    "comment_time": [faker.date_time_this_year()
                  for _ in range(100000)]
})
comments_df.head()
```

```python
comments_df.to_csv("raw_comments.csv",index=False)
```

```python
follows_df = pd.DataFrame({
    "follower_id": np.random.randint(1, n_users + 1, 50000),
    "followee_id": np.random.randint(1, n_users + 1, 50000),
    "follow_time": [faker.date_time_this_year()
                  for _ in range(50000)]
})
comments_df.head()
```

```python
follows_df.to_csv("raw_follows.csv",index=False)
```

```python
users_dirty = users_df.copy()
photos_dirty = photos_df.copy()
likes_dirty = likes_df.copy()
comments_dirty = comments_df.copy()
follows_dirty = follows_df.copy()
```

```python
idx = users_dirty.sample(frac=0.02, random_state=9).index
users_dirty.loc[idx, "user_name"] = np.nan
```

```python
idx = photos_dirty.sample(frac=0.01, random_state=9).index
photos_dirty.loc[idx, "upload_time"] = pd.NaT
```

```python
duplicate_likes= likes_dirty.sample(n=5000, random_state=9)
likes_dirty = pd.concat([likes_dirty, duplicate_likes], ignore_index = True)
```

```python
duplicate_comments = comments_dirty.sample(n=2000, random_state = 9)
comments_dirty = pd.concat([comments_dirty, duplicate_comments], ignore_index = True)
```

```python
idx = photos_dirty.sample(n=500, random_state = 9).index
photos_dirty.loc[idx, "upload_time"] ="2036-01-01"
```

```python
idx = follows_dirty.sample(n=1000, random_state = 9).index
follows_dirty.loc[idx, "followee_id"] = follows_dirty.loc[idx, "follower_id"]
```

```python
bot_users = users_dirty["user_id"].sample(n=5,random_state=9)
current_max = photos_dirty["photo_id"].max()
extra_photos=[]
for bot in bot_users:
    n_posts = random.randint(500,2000)
    for i in range(n_posts):
        current_max += 1
        extra_photos.append({
            "photo_id": current_max,
            "user_id": bot,
            "upload_time":faker.date_time_this_year()
        })
extra_photos_df = pd.DataFrame(extra_photos)
photos_dirty = pd.concat([photos_dirty, extra_photos_df],ignore_index =True)
```

```python
users_dirty.to_csv("users_dirty.csv",index=False)
photos_dirty.to_csv("photos_dirty.csv",index=False)
likes_dirty.to_csv("likes_dirty.csv",index=False)
comments_dirty.to_csv("comments_dirty.csv",index=False)
follows_dirty.to_csv("follows_dirty.csv",index=False)
```

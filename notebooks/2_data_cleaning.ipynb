```python
import pandas as pd
users_dirty = pd.read_csv("users_dirty.csv")
photos_dirty = pd.read_csv("photos_dirty.csv")
likes_dirty = pd.read_csv("likes_dirty.csv")
follows_dirty = pd.read_csv("follows_dirty.csv")
comments_dirty = pd.read_csv("comments_dirty.csv")
```

```python
users_dirty["register_date"] = pd.to_datetime(users_dirty["register_date"])
users_dirty.info()
users_dirty.describe(include='all')
```

```python
users_dirty.isnull().sum()
```

```python
users_clean = users_dirty.copy()
users_clean["user_name"] = users_clean["user_name"].fillna("unknown_user")
```

```python
photos_dirty["upload_time"] = pd.to_datetime(photos_dirty["upload_time"])
photos_dirty.info()
photos_dirty.describe(include='all')
```

```python
photos_clean = photos_dirty[photos_dirty["upload_time"] <= pd.Timestamp.today()]
```

```python
likes_dirty.info()
likes_dirty.describe(include="all")
```

```python
likes_dirty.duplicated(subset=["user_id","photo_id"]).sum()
```

```python
likes_clean = likes_dirty.drop_duplicates(subset=["user_id","photo_id"])
```

```python
follows_dirty.info()
follows_dirty.describe(include="all")
```

```python
follows_dirty.duplicated(subset=["follower_id","followee_id"]).sum()
```

```python
follows_clean = follows_dirty.drop_duplicates(subset=["follower_id","followee_id"])
follows_clean.duplicated(subset=["follower_id","followee_id"]).sum()
```

```python
comments_dirty.info()
comments_dirty.describe(include="all")
```

```python
comments_dirty.duplicated(subset=["comment_id","user_id"]).sum()
```

```python
comments_clean=comments_dirty.drop_duplicates(subset=["comment_id","user_id"])
comments_clean.duplicated(subset=["comment_id","user_id"]).sum()
```

```python
photos_user = photos_clean.merge(users_clean, on = "user_id", how= "left")
photos_user.describe(include="all")
```

```python
photos_user[photos_user["upload_time"]<photos_user["register_date"]]
```

```python
photos_user_clean = photos_user[photos_user["upload_time"] >= photos_user["register_date"]]
```

```python
user_posts = (photos_user_clean.groupby("user_id").size().reset_index(name="post"))
```

```python
user_posts.describe(include="all")
```

```python
Q1 = user_posts["post"].quantile(0.25)
Q3 = user_posts["post"].quantile(0.75)
IQR = Q3 - Q1
threshold = Q1 + 1.5 * IQR
```

```python
user_posts["is_abnormal"] = (user_posts["post"] > threshold)
user_posts[user_posts["is_abnormal"]]
```

```python
users_clean.to_csv("users_clean.csv", index=False)
photos_clean.to_csv("photos_clean.csv", index=False)
likes_clean.to_csv("likes_clean.csv", index=False)
follows_clean.to_csv("follows_clean.csv", index=False)
comments_clean.to_csv("comments_clean.csv", index=False)
```
